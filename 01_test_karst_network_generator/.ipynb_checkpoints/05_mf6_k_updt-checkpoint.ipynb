```python
#grid
nx = 100
ny = 100
sx = 4
sy = 4
sz = 4

#boundary condition
position_out = [1,0]
cstIn  = [[(0,ny-1,i), 100] for i in range(nx)]  #Charge cst top
cstOut = [[(0,position_out[0],position_out[1]), 0]]  #tunnel position
cst_bc = [cstIn, cstOut]

#conduit = 1 pour presence
conduit[karst_binaire==1] = karst_diameter[karst_binaire==1]
conduit = conduit
#conduit = np.zeros((ny,nx))
#conduit[conduit!=0] = 0.1

#matrix first iteration
Km = 1e-5
kmatrix_homogene = np.ones((ny,nx))*Km
kmatrix = assign_k_laminar(kmatrix_homogene, conduit/2, cell_size=sx, Km=Km)

#parameter nb run
diff_threshold   = 1e-5
nb_iteration_min = 5
nb_iteration_max = 15

#init variable
test    = False
outflow = 0
iteration = 0
diff_store    = []
km_store      = []
hd_store      = []
outflow_store = [outflow]

#run loop
while test is False or iteration<nb_iteration_min:
    #run Modflow
    run_all_mf6(kmatrix, nx, ny, sx, sy, sz, cst_bc, silent=True)
    
    #get outflow
    bdg         = fp.utils.Mf6ListBudget('file_modflow_6/model_mf6.lst')    
    outflow_new = bdg.get_budget()[0]['TOTAL_OUT'][0]
    diff_store.append(np.abs((outflow-outflow_new)/outflow_new))
    outflow_store.append(outflow_new)
    outflow     = outflow_new
    
    #get outputs
    cbc_flux = get_spdis('file_modflow_6/model_mf6.cbc')[0]
    velocity = get_velocity(cbc_flux, conduit/2)
    hds      = get_hds('file_modflow_6/model_mf6.hds')
    grad_h   = get_gradient_y(hds, conduit, sx=sx, sy=sy)
    hd_store.append(hds)
    
    #calculate Re
    rey_mat, rey_test = reynolds(velocity, conduit)
    
    #update k matrix    
    km_store.append(np.copy(kmatrix))
    k_t      = kturb(conduit/2, cell_size=sx, Km=Km, gradh=grad_h)
    kmatrix  = update_km(kmatrix, k_t)
    
    #test convergence
    if diff_store[iteration]<diff_threshold:
        test = True
    iteration +=1 
    
    if iteration == nb_iteration_max:
        break

#print output information        
print('Number of iteration : {}'.format(iteration))
print('Final budget out : {}m3/s'.format((outflow)))
print('Final diff in out flow : {}m3/s'.format(diff_store[-1]))
```
